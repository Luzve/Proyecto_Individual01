```python
netflix=pd.read_csv(r'../Datasets/amazon_prime_titles.csv', encoding = 'utf-8')
netflix.head()
```

```python
#Se realiza una  copia a los dataframe
amazon_df=amazon.copy()
disney_df=disney.copy()
hulu_df=hulu.copy()
netflix_df=netflix.copy()
```

```python
#Se agrega una columna a cada dataframe para asignarle la ´plataforma que le corresponde

amazon_df['Plataforma']='amazon'
disney_df['Plataforma']='disney'
hulu_df['Plataforma']='hulu'
netflix_df['Plataforma']='netflix'
```

```python
#Reemplazar campo vacios en la columna duration
hulu_df['duration'].fillna('sin datos',inplace=True)
netflix_df['duration'].fillna('sin datos',inplace=True)
```

```python
#Función que divide la columna duration
def sepdurationcolum(df):
    df['temporadas']=df['duration'].apply( lambda i: i.split()[0] if i.split()[1]=='Seasons' else '0')
    df['duracion_peliculas']=df['duration'].apply( lambda i: i.split()[0] if i.split(" ",1)[1]=='min' or i.split(" ",1)[1]=='min' else '0')
    return df
```

```python
#Se implementa la función anterior en cada dataset
amazon_df=sepdurationcolum(amazon_df)
disney_df=sepdurationcolum(disney_df)
hulu_df=sepdurationcolum(hulu_df)
netflix_df=sepdurationcolum(netflix_df)
```

```python
#Elimina la columna duration en cada dataset
amazon_df.drop(columns='duration', inplace=True)
disney_df.drop(columns='duration', inplace=True)
hulu_df.drop(columns='duration', inplace=True)
netflix_df.drop(columns='duration', inplace=True)
```

```python
#concatenar todos los dataframe
final_df=pd.concat([amazon_df, disney_df, hulu_df, netflix_df], axis=0)
```

```python
#Verificar tipo de dato de cada columna
final_df.info()
```

```python
#Transformacion del tipo de datos
final_df.duracion_peliculas =final_df.duracion_peliculas.astype(int)
```
